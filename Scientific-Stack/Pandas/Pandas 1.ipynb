#### Groupby computations

```python
# shows number of rows per group
oo.groupby('Edition').size()
```

#### agg([..])

```python
oo.groupby(['Edition', 'NOC', 'Medal']).agg(['min', 'max', 'count'])
```

```python
oo.groupby(['Edition', 'NOC', 'Medal']).size()
```

#### agg({:[...]})

```python
# static output for min max bc they're string type
oo.groupby(['Edition', 'NOC', 'Medal']).agg({'Edition': ['min', 'max', 'count']}) 
```

```python
# min => first olymp, max => last, count => total medals
oo.loc[oo.Athlete == 'LEWIS, Carl'].groupby('Athlete').agg({'Edition': ['min', 'max', 'count']}) 
```

### Module 8

#### Reshaping

```python
import pandas as pd
```

```python
# reading in data
oo = pd.read_csv('data/olympics.csv', skiprows=4)
oo.head()
```

```python
# gathering data for 2008 100m and 200m
mw = oo[(oo.Edition == 2008) & ( (oo.Event == '100m') | (oo.Event == '200m') )]
mw
```

```python
g = mw.groupby(['NOC', 'Gender', 'Discipline', 'Event']).size()
g
```

#### unstack

```python
# long to wide
df = g.unstack(['Discipline', 'Event']) # moves these categories to column headers
df
```

#### stack

```python
# stack: goes wide to long
df.stack()
```

```python
# specify which column you're stacking
df.stack('Event')
```

```python
# unstack: goes long to wide
df.unstack()
```

```python
# prefer to specify which column to unstack
df.unstack('Gender') # typically multi level headers here
```

```python
# gold medal usa, plot by gender over the years
gu = oo[(oo.NOC == 'USA') & (oo.Medal == 'Gold')]
gu.groupby(['Edition', 'Gender']).size().unstack('Gender', fill_value=0).plot()
```

```python
g = oo.groupby(['Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
g.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']].head().plot(kind='bar')
```

```python
# 
gy = oo[oo.NOC == 'USA']
gy = gy.groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']
gy.reset_index(inplace=True)
# in a list comprehension, the action comes firt, then the iterator, all enclosed in square brackets
tu = [group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby('Edition')] # list comprehension
tu
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
